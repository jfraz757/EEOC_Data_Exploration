In my previous data cleaning, I overlooked that the first row of my data set for each new year, shows an aggregated total for each value. I need to remove that aggregated total value. 

```python
#find the rows that are the first index for each year. These rows should contain the aggregated values in the columns.
#There should be 1 row for each year, totaling 10 rows
first_index_yr = df.groupby('Year').apply(lambda x: x.index[0])
first_index_yr.head(11)
```

```python
#Double checking to see if these are the correct rows
df.iloc[[0, 43707, 87886, 132346, 176074, 219993, 264325, 307140, 352283, 455758]]
```

```python
#Dropping those Specific aggregated rows
droped_ag_rows = df.drop(index=first_index_yr)
```

```python
#testing to see if those aggrgated values were dropped from the dataset
droped_ag_rows.head()
```

```python
#Ensure that the totals for the columns in the new dataframes equal the aggregated columns that were dropped
droped_ag_rows.groupby("Year")["Male_Total"].sum()
```

The numbers above are higher than they should be! Compare those to the numbers below, which represent what the dropped aggregated values were originally in the data set.
