```python
#previous code still did not work. Trying this AI corrected version instead. 
# 1. Grab the first 11 columns (leave them exactly as they are)
left_side = df.iloc[:, :11]

# 2. Grab column 11 onwards and convert them to numbers
right_side = df.iloc[:, 11:].apply(pd.to_numeric, errors='coerce')

# 3. Glue them back together side-by-side (axis=1) into a new dataframe
df = pd.concat([left_side, right_side], axis=1)

df.dtypes
df.info()
```

```python
#Looking to see if Louisville, KY is in the 'CBSA' column.
# This returns a DataFrame containing only rows where the CBSA name includes "Louisville"
louisville_data = df[df["CBSA"].str.contains("Louisville", case=False, na=False)]

# To see the specific counts for those matching names:
print(louisville_data["CBSA"].value_counts())
```

```python
# 1. Isolate the national totals (Where State is blank)
df_national_baselines = df[df['STATE'].isna()].copy()

# 2. Isolate the State-level totals (Where State is NOT blank, but County is blank)
df_state_totals = df[df['STATE'].notna() & df['COUNTY'].isna()].copy()

# 3. Create your main granular dataset (Where County is NOT blank)
df_county_totals = df[df['COUNTY'].notna()].copy()

# 4. Create your main granular dataset (Where CBSA is NOT blank)
df_CBSA_total = df[df['COUNTY'].notna()].copy()
```
